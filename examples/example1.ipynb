```python
import networkx as nx
import network_analysis.analysis as network
import network_analysis.dismantling as network_dismantling

G = nx.petersen_graph()
N = network.NetworkAnalysis(G)

N.draw(layout=nx.nx_agraph.graphviz_layout, node_size=100, with_labels=False)

print(N.degree_centrality())
print(N.betweenness_centrality())
print(N.eigenvector_centrality())
print(N.closeness_centrality())
print(N.articulation_points())
print(N.global_efficiency())
print(N.largest_connected_component())
```
